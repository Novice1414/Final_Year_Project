NOW WE WILL USE CONVOLUTIONAL NEURAL NETWORKS TO IMPROVE THE ACCURACY OF THE MODEL

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Building a 1D Convolutional Neural Network (Conv1D) model with MaxPooling1D
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(8,1)),

    # First Conv1D Layer
    tf.keras.layers.Conv1D(64, kernel_size=3, activation='ReLU'),



      tf.keras.layers.MaxPooling1D(pool_size=1),

    #Third Convolution Layer
    tf.keras.layers.Conv1D(256, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=2),


    # Flatten the output for dense layers
    tf.keras.layers.Flatten(),


    # Fully connected layers
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),

    # Output layer
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)  # You can adjust the learning rate
model.compile(optimizer=sgd_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```
